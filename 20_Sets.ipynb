# Set Operations

### union

```python
f = {1,2,3,4,5,6}
s = {4,5,6,7,8,9}

print(f | s)  # union
```

### intersection

```python
f = {1,2,3,4,5,6}
s = {4,5,6,7,8,9}

print(f & s)  # intersection
```

### difference

```python
f = {1,2,3,4,5,6}
s = {4,5,6,7,8,9}

print(f - s)  # difference
```

```python
print(s - f)
```

### symmetric

```python
f = {1,2,3,4,5,6}
s = {4,5,6,7,8,9}

print(f ^ s)  # symmetric
```

# set functions


```python
n = {11,22,3,4,5,47}

print(len(n))
print(max(n))
print(min(n))
print(sum(n))
```
