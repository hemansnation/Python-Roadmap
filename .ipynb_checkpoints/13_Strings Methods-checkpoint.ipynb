**WAP to enter a string from user and print reverse of it without slicing**

```python
s = input("Enter a string")

#s = "indore"
#     012345
n = len(s) - 1

while n >= 0:
    print(s[n], end="")
    n -= 1
```

### Join

```python
s = "+"

f = s.join(['A', 'b', 'C'])

print(f)
print(type(f))
```

### replace

```python
s = "this is virat kohli. He also known as king kohli"

a = s.replace("kohli" , "sharma")
print(a)
```

```python
s = "this is virat kohli. He also known as king kohli"

a = s.replace("kohli" , "sharma", 1)
print(a)
```

### split

```python
s = "Hello World"

l = s.split(" ")

print(l)
```

```python
s = "Hello World"

l = s.split("W")

print(l)
```

```python
s = "Hello World"

l = s.split("o")

print(l)
```

```python
s = "1+2+3+4+5"

l = s.split("+")

print(l)
```

```python
print(sum(l))
```

```python
s = "1+2+3+4+5"

l = s.split("+")
print(l)
i = 0
n = 0

while i < len(l):
    n = n + int(l[i])
    i += 1
print(n)
```
