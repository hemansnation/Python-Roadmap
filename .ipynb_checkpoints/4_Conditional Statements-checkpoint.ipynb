### Nested If else

```python
a = 7

if a > 4:
    print(1)
    if a == 5:
        print(2)
    else:
        print(3)
        if a == 7:
            print(4)
else:
    print(5)
```

```python
a = int(input("Enter a number: "))

if a > 4:
    print(1)
    if a == 5:
        print(2)
        if a < 3:
            print(3)
        elif a > 4:
            print(4)
    elif a == 7:
        print(5)
        if a != 7:
            print(6)
        else:
            print(7)
else:
    if a == 1:
        print(8)
    else:
        print(9)
```
