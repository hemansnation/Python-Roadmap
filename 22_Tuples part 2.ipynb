# Tuple Unpacking

```python
n = (1,2,3)

a,b,c = n

print(a)
print(b)
print(c)
```

```python
n = (1,2,3)

a,b,c,d = n

print(a)
print(b)
print(c)
```

```python
n = (1,2,3,4)

a,b,c = n

print(a)
print(b)
print(c)
```

## Swap two numbers

```python
a = 10
b = 20

c = a
a = b
b = c

print(a)
print(b)
```

### without using third variable

```python
a = 10  # a = 10, 30, 20
b = 20  # b = 20, 10

a = a + b
b = a - b
a = a - b

print(a)
print(b)
```

### swapping using tuple unpacking

```python
a = 10
b = 20

a,b = b,a
#a,b = (20,10)

print(a)
print(b)
```

#### list unpacking

```python
a = 10
b = 20

a,b = [b,a]
#a,b = (20,10)

print(a)
print(b)
```

## Tuple Slicing

```python
#                                -1
t = (11,22,33,44,55,66,77,88,99,100)
#    0  1
print(t)
```

```python
print(t[3:5])
```

```python
print(t[:])
```

```python
print(t[-8:-4])
```

```python
print(t[-4:-8])
```

```python
print(t[-4:-8:-1])
```

```python
print(t[::-1])
```

```python
print(t[::-2])
```
